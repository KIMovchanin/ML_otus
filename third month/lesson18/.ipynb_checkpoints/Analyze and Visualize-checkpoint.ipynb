### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
import numpy as np

nums = np.zeros(10)
nums[4] = 1
nums = nums.reshape(2, 5)

print(nums)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
import numpy as np

nums = np.arange(10, 50)
nums = nums[::-1]
chet = [i for i in nums if i % 2 == 0]

for i in chet:
    print(i)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
import numpy as np

np.random.seed(7)

nums = np.random.randint(-1000, 1000, size=(4, 3, 2))

min_num = 0
max_num = 0

for cube in range(len(nums)):
    for row in range(len(nums[cube])):
        for column in range(len(nums[cube][row])):
            if min_num > nums[cube][row][column]:
                min_num = nums[cube][row][column]
            if max_num < nums[cube][row][column]:
                max_num = nums[cube][row][column]

print(nums)
print(f"Минимальное число: {min_num}. Максимальное число: {max_num}.")
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
import numpy as np

np.random.seed(7)

arr1 = np.random.randint(-100, 100, size=(6, 4))
arr2 = np.random.randint(-100, 100, size=(4, 3))

print(arr1 @ arr2) # Либо print(np.dot(arr1, arr2))
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
import numpy as np

np.random.seed(7)

nums = np.random.randint(-100, 100, size=(7, 7))

mean = np.mean(nums)
std = np.std(nums)

print("Среднее значение:", mean)
print("Стандартное отклонение:", std)

normalized_arr = (nums - mean) / std

print("Нормализованный массив:")
print(normalized_arr)
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
import numpy as np
import pandas, matplotlib.pyplot as plt, seaborn

df = seaborn.load_dataset('tips')

plt.figure(figsize=(8, 6))
plt.hist(df['total_bill'], bins=20, color='skyblue', edgecolor='black')

plt.title('Гистограмма распределения признака "total_bill"')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')

plt.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')

sns.scatterplot(x='total_bill', y='tip', data=df)

plt.title('Взаимосвязь между total_bill и tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

plt.show()
```

### Постройте pairplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')

sns.pairplot(df)

plt.show()
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')

sns.boxplot(x='day', y='total_bill', data=df)

plt.title('Взаимосвязь между total_bill и day')
plt.xlabel('Day')
plt.ylabel('Total Bill')

plt.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')

sns.histplot(data=df, x='tip', hue='time', kde=True, multiple='stack')
plt.title('Распределение tip по категориям time')
plt.xlabel('Tip')
plt.ylabel('Frequency')

plt.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x='total_bill', y='tip', data=df[df['sex'] == 'Male'], hue='smoker')
plt.title('Взаимосвязь total_bill и tip для Male')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

plt.subplot(1, 2, 2)
sns.scatterplot(x='total_bill', y='tip', data=df[df['sex'] == 'Female'], hue='smoker')
plt.title('Взаимосвязь total_bill и tip для Female')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

plt.tight_layout()
plt.show()
```
